```python
import pandas as pd
import matplotlib.pyplot as plt

iid_dicts = [kitti_iid_4, kitti_iid_10, foggy_iid_4, cityscapes_iid_4, sim10k_iid_4]
noniid_dicts = [kitti_noniid_4, kitti_noniid_10, foggy_noniid_4, cityscapes_noniid_4, sim10k_noniid_4]
titles = ['kitti_4', 'kitti_10', 'foggy_4', 'cityscapes_4', 'sim10k_4']

fig, axes = plt.subplots(2, 5, figsize=(25, 10))

for i in range(5):
    # iid row: 不顯示 legend
    dict_data = iid_dicts[i]
    client_count = len(next(iter(dict_data.values())))
    clients = [f"C{j+1}" for j in range(client_count)]
    df = pd.DataFrame(dict_data, index=clients)
    df.plot.barh(stacked=True, ax=axes[0, i], legend=False)
    axes[0, i].set_title(titles[i] + ' (iid)', fontsize=14)
    axes[0, i].set_xlabel('Samples', fontsize=14)
    axes[0, i].set_ylabel('Client', fontsize=14)

    # noniid row: 0, 1 legend右下角，其餘右上
    dict_data_noniid = noniid_dicts[i]
    client_count_noniid = len(next(iter(dict_data_noniid.values())))
    clients_noniid = [f"C{j+1}" for j in range(client_count_noniid)]
    df_noniid = pd.DataFrame(dict_data_noniid, index=clients_noniid)
    df_noniid.plot.barh(stacked=True, ax=axes[1, i], legend=False)
    axes[1, i].set_title(titles[i] + ' (non-iid)', fontsize=14)
    axes[1, i].set_xlabel('Samples', fontsize=14)
    axes[1, i].set_ylabel('Client', fontsize=14)
    handles, labels = axes[1, i].get_legend_handles_labels()
    if i in [0, 1]:
        axes[1, i].legend(handles, labels, loc='lower right', fontsize=14, frameon=True)
    else:
        axes[1, i].legend(handles, labels, loc='upper right', fontsize=14, frameon=True)

plt.tight_layout()
plt.show()
```
