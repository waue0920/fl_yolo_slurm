```python
import wandb

def truncate_wandb_media(entity, project):
    api = wandb.Api()
    runs = api.runs(f"{entity}/{project}")
    for run in runs:
        print(f"Processing run: {run.id}")
        for file in run.files():
            if file.name.startswith("media/"):
                # 若只刪圖片，可再加副檔名條件
                if file.name.endswith(".png") or file.name.endswith(".jpg"):
                    print("Deleting:", file.name)
                    file.delete()
                # 若要刪所有 media/ 下檔案，則移除副檔名條件。
    print("Media cleanup finished.")


```

```python
# 使用範例
truncate_wandb_media("nchc", "63_kittiO_fednova_10C_12R_202511042253")
```
